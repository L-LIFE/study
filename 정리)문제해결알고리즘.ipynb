# 동전

```python
import random

print(random.choice("앞뒤")) # 인덱스 0: 앞 / 인덱스 1: 뒤

#Q. "앞면", "뒷면" 중에서 선택하는 건?
print(random.choice("앞면뒷면")) # 이렇게 작성하면 => 인덱스 0: 앞/ 1:면/ 2:뒷/ 3:면 으로 저장됨

# 이와 같은 방식으로 진행되어야 한다.
ch=["앞면", "뒷면"] # 값을 여러개 저장할 수 있는 리스트
print(random.choice(ch)) # Q에 대한 원하는 결과가 나온다.
```

### 문제

1. di1과 di2변수를 만들고, 주사위 값 2개를 각각 생성하여 저장.
2. 이 주사위 값이 무엇이 나왔는지 두 주사위 값의 합은 무엇인지를 구한다.
3. 나온 눈은 (di1값)과 (di2값)이며, 두 주사위의 합은 (di1+di2)입니다.
4. 형식으로 출력하기, ()안의 내용은 변수값을 활용할 것.

```python
import random

di1= random.randrange(1,7)
di2= random.randrange(1,7)

#print문으로 작성할 시
print("나온 눈은", di1, "과", di2,"이며, 두 주사위의 합은",di1+di2,"입니다.")

#format 함수 사용할 시
print(f"나온 눈은 {di1} 과 {di2} 이며, 두 주사위의 합은 {di1+di2} 입니다.")
print("나온 눈은 {0} 과 {1} 이며, 두 주사위의 합은 {2} 입니다.".format(di1,di2,di1+di2))
```

#시간 라이브러리

```python
import datetime

today= datetime.datetime.now()

print("지금은 ",today.year,"년입니다.")
print("지금은 ",today.month,"월입니다.")
print("지금은 ",today.day,"일입니다.")
print("지금은 ",today.weekday(),"요일입니다.")
print("지금은 ",today.hour,"시간입니다.")
print("지금은 ",today.minute,"분입니다.")
```

### weekday()

weekday(): 결과를 숫자로 받은 경우 0~6까지의 숫자를 받을 수 있다.  

0:월요일  
1:화요일  
2:수요일  
3:목요일  
4:금요일  
5:토요일  
6:일요일  

숫자로 요일을 알 수 있다.

```python
import datetime

if today.weekday()==6:
  print("오늘은 일요일입니다.")
elif today.weekday()==5:
  print("오늘은 토요일입니다.")
elif today.weekday()==4:
  print("오늘은 금요일입니다.")
elif today.weekday()==3:
  print("오늘은 목요일입니다.")
elif today.weekday()==2:
  print("오늘은 수요일입니다.")
elif today.weekday()==1:
  print("오늘은 화요일입니다.")
else:
  print("오늘은 월요일입니다.")
```

#or연산자와 in연산자

or 보다 in 연산자를 사용할 시 좀 더 짧은 코드를 작성할 수 있다.

```python
#or연산자인 경우

import datetime

today=datetime.datetime.now()

if today.weekday()==5 or today.weekday()==6:
  print("오늘은 주말입니다.")
else:
  print("오늘은 평일입니다.")
```

```python
#in연산자인 경우

import datetime

today=datetime.datetime.now()

if today.weekday() in [5,6]:
  print("오늘은 주말입니다.")
else:
  print("오늘은 평일입니다.")
```

#문자 변환: 대문자, 소문자, 공백 제거

```python
name=" Ggm "

print("변환하기 전")
print(name)
print("전체 대문자")
print(name.upper())
print("전체 소문자")
print(name.lower())
print("전체 공백 제거")
print(name.strip())
print("오른쪽 공백 제거")
print(name.rstrip())
print("왼쪽 공백 제거")
print(name.lstrip())
print("전체 대문자& 전체 공백 제거")
print(name.upper().strip())
```
