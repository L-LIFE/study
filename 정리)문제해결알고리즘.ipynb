### 문제

1. di1과 di2변수를 만들고, 주사위 값 2개를 각각 생성하여 저장.
2. 이 주사위 값이 무엇이 나왔는지 두 주사위 값의 합은 무엇인지를 구한다.
3. 나온 눈은 (di1값)과 (di2값)이며, 두 주사위의 합은 (di1+di2)입니다.
4. 형식으로 출력하기, ()안의 내용은 변수값을 활용할 것.

```python
import random

di1= random.randrange(1,7)
di2= random.randrange(1,7)

#print문으로 작성할 시
print("나온 눈은", di1, "과", di2,"이며, 두 주사위의 합은",di1+di2,"입니다.")

#format 함수 사용할 시
print(f"나온 눈은 {di1} 과 {di2} 이며, 두 주사위의 합은 {di1+di2} 입니다.")
print("나온 눈은 {0} 과 {1} 이며, 두 주사위의 합은 {2} 입니다.".format(di1,di2,di1+di2))
```

#시간 라이브러리

```python
import datetime

today= datetime.datetime.now()

print("지금은 ",today.year,"년입니다.")
print("지금은 ",today.month,"월입니다.")
print("지금은 ",today.day,"일입니다.")
print("지금은 ",today.weekday(),"요일입니다.")
print("지금은 ",today.hour,"시간입니다.")
print("지금은 ",today.minute,"분입니다.")
```

### weekday()

weekday(): 결과를 숫자로 받은 경우 0~6까지의 숫자를 받을 수 있다.  

0:월요일  
1:화요일  
2:수요일  
3:목요일  
4:금요일  
5:토요일  
6:일요일  

숫자로 요일을 알 수 있다.

```python
import datetime

if today.weekday()==6:
  print("오늘은 일요일입니다.")
elif today.weekday()==5:
  print("오늘은 토요일입니다.")
elif today.weekday()==4:
  print("오늘은 금요일입니다.")
elif today.weekday()==3:
  print("오늘은 목요일입니다.")
elif today.weekday()==2:
  print("오늘은 수요일입니다.")
elif today.weekday()==1:
  print("오늘은 화요일입니다.")
else:
  print("오늘은 월요일입니다.")
```

#or연산자와 in연산자

or 보다 in 연산자를 사용할 시 좀 더 짧은 코드를 작성할 수 있다.

```python
#or연산자인 경우

import datetime

today=datetime.datetime.now()

if today.weekday()==5 or today.weekday()==6:
  print("오늘은 주말입니다.")
else:
  print("오늘은 평일입니다.")
```

```python
#in연산자인 경우

import datetime

today=datetime.datetime.now()

if today.weekday() in [5,6]:
  print("오늘은 주말입니다.")
else:
  print("오늘은 평일입니다.")
```

#가위바위보 게임

```python
import random

print("가위바위보 게임을 시작합니다.")
Uchoice=input("가위(S), 바위(R), 보(P)중 하나만 입력하세요: " )

Cchoice=random.choice("SRP")

Uchoice=Uchoice.upper().strip()

print("사용자는 ", Uchoice)
print("컴퓨터는 ", Cchoice)


if Uchoice==Cchoice:
  print("무승부")
elif Uchoice=="S" and Cchoice=="P":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 승리!")
elif Uchoice=="R" and Cchoice=="S":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 승리!")
elif Uchoice=="P" and Cchoice=="R":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 승리!")
elif Uchoice=="S" and Cchoice=="R":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 패배!")
elif Uchoice=="R" and Cchoice=="P":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 패배!")
elif Uchoice=="P" and Cchoice=="S":
  print("사용자는 ",Uchoice, "컴퓨터는", Cchoice, "사용자가 패배!")
else:
  print("다른 값을 입력하였으니 게임을 종료합니다.")
```

```python
animals=["강아지","고양이","친칠라","개미","공룡"]

print(animals)

print("animals 리스트 안에는 ", len(animals),"마리의 동물이 들어가 있습니다.")

#요소 하나씩 불러와 출력도 가능
print("animals의 첫번째 위치에는 ", animals[0],"가 들어가 있습니다.")
print("animals의 네번째 위치에는 ", animals[3],"가 들어가 있습니다.")

print("기존 animals의 2번방 값은 ",animals[2])
animals[2]="염소"
print("변경된 animals의 2번방 값은 ",animals[2])
```

```python
#추가, 병합, 삭제
am=["강아지","고양이","병아리"]
print(am)

#요소 추가하기=> append()
am.append("양")
print(am)

am2=["염소","말"]

#am 리스트에 am2리스트의 내용을 병합
am.extend(am2)
print(am)

#요소 삭제=> pop()위치 기반(인덱스 0을 지우고 싶어!), remove() 데이터 기반 삭제(염소를 지우고 싶어!)
#염소 삭제=>염소 위치는 ? am[4]
am.pop(4)
print(am)
am.remove("고양이")
print(am)
```

```python
#단어의 개수와 오름차순 => len() , sort(), sort(reverse=True)

am=['강아지','고양이','오리','하마','도마뱀']

print("동물은 ",len(am),"마리가 있습니다.")

print("오름차순 적용 전: ",am)
am.sort()
print("오름차순 적용 후: ",am)

am.sort(reverse=True)
print("내림차순 적용 후: ",am)
```

```python
#1부터 10까지 반복문을 통해 합을 구해본다

sum=0

for i in range(1,11):
  sum=sum+i
  print("i의 값은: ",i,"이고, sum의 누적된 값은: ", sum, "이다.")
```

```python
#중첩 반복문 (중첩 for문)을 사용한 구구단 출력하기

for i in range(2,10):
  print("="*8,"구구단 ",i,"단 출력","="*8)
  for j in range(1,10):
    print(i,"*",j,"=",i*j)
```
