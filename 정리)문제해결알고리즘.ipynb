```python
# 암호키 : 314159

#min, max를 저장한 이유가 아스키코드 범위(32~126를 벗어나지 말라는 의미.
acmin=32 #공백을 아스키코드로 변환한 값
acmax=126 # ~를 아스키코드로 변환한 값

key=314159 #제공된 암호키
key=str(key) #문자로 접근하기 위해 암호키의 타입을 바꾼다.

msg= input("암호화할 메시지를 입력하세요.")

msgec=""

for index in range(0,len(msg)):
  char=ord(msg[index])
  if char<acmin or char>acmax:
    msgec+=msg[index]
  else:
    acnum=char+int(key[index%len(key)])
    #암호화 진행=>
    #index%len(key) : 암호키 314159에서 몇번째를 적용하면 좋을지 알아내기 위해
    #만약 10번째 인덱스가 암호화하기 위해서 어떤 숫자를 더해야 하는지 모르면/ 암호화 자리수를 나누면 몇번째 숫자를 쓰면 될지 알 수 있다.
    #ex) 인덱스 10 /암호키 6= 5번째 암호키 쓰면 됨=> 숫자 5를 더하면 됨.
    if acnum>acmax:
      acnum-=(acmax-acmin)
    msgec=msgec+chr(acnum)
print("암호화된 메시지: ",msgec)

```

#복호화

```python
acmin= 32
acmax=126

key= 314159
key= str(key)

msg= input("복호화할 메시지를 입력하세요")

msgec= ""

for index in range(0,len(msg)):
  char= ord(msg[index])
  if char < acmin or char> acmax:
    msgec += msg[index]
  else:
    acnum= char-int(key[index%len(key)])
    if acnum<acmin:
      acnum+=(acmax-acmin)
    msgec=msgec+chr(acnum)

print("암호화된 메시지: ", msgec)
```

#숫자 맞히기 게임 퀴즈

```python
#랜덤 라이브러리 불러오기
import random

#사용자 입력(문자)로 전달받아 저장할 userIP 를 빈칸으로 초기화
userIP=""

#사용자 입력을 숫자로 저장할 공간 userGS초기화
userGS=0

#1부터 100까지 무작위 숫자를 선정하여 rnm에 저장
rnm=random.randrange(1,101)

print("1과 100 사이의 숫자를 하나 뽑았어요! 이 숫자는 무엇일까요?")

#랜덤으로 선정된 숫자와 사용자가 입력한 값인 userGS값이 다른동안 반복 (맞힐 때 까지 반복)
while rnm!=userGS :

#반복되는동안
#1. 사용자에게 답을 입력받아 공백을 제거하고 userIP에 저장
  userIP= input("1부터 100사이의 숫자를 입력해주세요: ").strip()

#2. 아래 문장 추가
  if not userIP.isnumeric():
#3. 2번의 조건문에 만족하면, 사용자에게 userIP의 값과 함께 해당값을 숫자가 아닙니다 출력.
    print("userIP값은 ",userIP,"이며, 해당하는 숫자가 아닙니다.")
#4. 2번의 조건문에 만족하지 않으면, userGS에 userIP값을 정수로변환하여 저장하고,
  else:
    userGS= int(userIP)
#5. 다시 조건문으로 질의
#   5-1. userGS의 값과 rnm의 값을 비교하였을떄, userGS의 값이 더 작으면, 너무 작습니다 출력
  if userGS<rnm:
    print("너무 작습니다.")
#   5-2. userGS의 값과 rnm의 값을 비교하였을떄, userGS의 값이 더 크면, 너무 큽니다 출력
  elif userGS>rnm:
    print("너무 큽니다.")
#   5-3. 아니면 정답입니다 출력.
  else:
    print("정답입니다.")
#조건문에서 벗어나 게임을 종료합니다. 출력
print("게임을 종료합니다.")
```

# 몇번만에 맞췄는 지 확인

```python
import random

userIP=""

userGS=0

num=0

rnm=random.randrange(1,101)

print("1과 100 사이의 숫자를 하나 뽑았어요! 이 숫자는 무엇일까요?")

while rnm!=userGS :
  userIP= input("1부터 100사이의 숫자를 입력해주세요: ").strip()
  if not userIP.isnumeric():
    print("userIP값은 ",userIP,"이며, 해당하는 숫자가 아닙니다.")
  else:
    num+=1 #이쪽에 추가하는 이유는 else문이 실행되야 숫자 비교가 시작되기 때문.
    userGS= int(userIP)
    if userGS<rnm:
      print("너무 작습니다.")
    elif userGS>rnm:
      print("너무 큽니다.")
    else:
      print("정답입니다.")
print(num,"번만에 성공했습니다.")
print("게임을 종료합니다.")
```

#무작위 숫자를 nmin과 nmax 사이값으로 선정하도록 변경.  

```python
import random

userIP=""

userGS=0

GS=0

nmin=1
nmax=100

rnm=random.randrange(nmin, nmax+1)

print("1과 100 사이의 숫자를 하나 뽑았어요! 이 숫자는 무엇일까요?")

while rnm!=userGS :
  userIP= input("1부터 100사이의 숫자를 입력해주세요: ").strip()
  if not userIP.isnumeric():
    print("userIP값은 ",userIP,"이며, 해당하는 숫자가 아닙니다.")
  else:
    GS+=1
    userGS= int(userIP)
    if userGS<nmin or userGS>nmax:
      print("범위에 해당하는 값이 아닙니다.")
    elif userGS<rnm:
      print("너무 작습니다.")
    elif userGS>rnm:
      print("너무 큽니다.")
    else:
      print("정답입니다.",GS,"번만에 성공했습니다.")
print("게임을 종료합니다.")
```
