```python
class Calc:
  def __init__(self, init_value):
    self.value=init_value
  def add(self, n):
    return self.value+ n
  def sub(self, n):
    return self.value- n
  def mul(self, n):
    return self.value* n
  def div(self, n):
    return self.value/ n

cal= Calc(100)
print("value={0}".format(cal.value))

a= cal.add(100)
b= cal.sub(50)
c= cal.mul(2)
d= cal.div(2)
print("a= {0}, b= {1}, c= {2}, d= {3}". format(a, b, c, d))
```

### 원핫 인코딩 : 텍스트나 범주형 변수를 수치형 변수로 변환할 수 있다.  

단점: 불필요한 저장 공간 차지, 단어의 의미를 저장할 수 없음, 유사한 단어의 유사성을 모름  

해결방안: 임베딩  

단어를 벡터로 나타낼 수 있다.

```python
from konlpy.tag import Komoran
import numpy as np

komoran= Komoran()
text= "오늘 날씨는 구름이 많아요"

nouns= komoran.nouns(text)
print(nouns)
dics={}

for word in nouns:
  if word not in dics.keys():
    dics[word]= len(dics)  #딕셔너리에 새로운 단어를 추가하기 전에 해당 딕셔너리의 키의 개수를 반환하는 것.
    #간단하게 생각하면 key와 value를 연관지어서 생각하면 될 거 같다.
print(dics)

#원-핫 인코딩 진행
nc= len(dics) #원-핫 인코딩을 진행할 단어의 개수
print(nc)
target= list(dics.values()) #각 단어의 고유한 인덱스를 리스트로 변환.
onehot= np.eye(nc)[target]
#원-핫 인코딩 진행
#np.eye() : 단위 행렬을 생성하는 함수.
#즉, nc × nc 크기의 단위 행렬을 만들고, 단위 행렬에서 target에 저장된 인덱스에 해당하는 행만 선택한다.

print(onehot)
```

###  Word2Vec 모델 학습

```python
from gensim.models import Word2Vec
from konlpy.tag import Komoran
import time

def read_review_data(filename):
  with open(filename, 'r') as f:
    data= [line.split('\t') for line in f.read().splitlines()]
    #파일의 내용을 라인별로 읽어와 분리하여 line에 순서대로 저장-> line에 저장된 한줄을 tab으로 분리한다.
    data= data[1:] #첫 번째 라인인  헤더를 제외한 데이터를 반환.
  return data

start= time.time() #작업 시간으로 기록할 때 사용

print("1) 말뭉치 데이터 읽기 시작!")
review_data= read_review_data("./ratings.txt") #read_review_data(filename) 함수를 사용한 데이터 정리
#리뷰 데이터 개수
print("리뷰 데이터 개수", len(review_data))
#time.time()-start=> 현재 시간- 시작한 시간 = 특정 작업이 시작한 후 결과한 시간을 초단위로 나타낸 값
print("1) 말뭉치 데이터 읽기 완료: ", time.time()-start)
print("-"*30)

#명사만 추출
print("2) 형태소에서 명사만 추출 시작")
komoran=Komoran()
#sentence[1] : ratings.txt에는 1열에 리뷰에 관한 데이터가 들어있음.
docs= [komoran.nouns(sentence[1]) for sentence in review_data]
print("2) 형태소에서 명사만 추출 완료 ", time.time()-start)
print("-"*30)

#일반적으로 모델을 학습할 때 명사만 추출하여 학습시키는 것이 일반적이다.
#텍스트 데이터에 대한 전처리 과정 중 명사만 추출하는 게 있음.

print("3) word2vec 모델 학습 시작")
model=Word2Vec(sentences=docs, vector_size=200, window=4, min_count=2, sg=1)
print("3) word2vec 모델 학습 완료 ", time.time()-start)
print("-"*30)

print("4) 학습된 모델 저장 시작")
model.save("nvmx.model")
print("4) 학습된 모델 저장 완료", time.time()-start)
print("-"*30)

#학습된 말뭉치 개수
print("corpus_count(문장의 개수): ",model.corpus_count)
print("corpus_total_words(문장의 전체 단어 개수): ", model.corpus_total_words)

```

### 학습된 Word2Vec모델 로드

학습된 모델을 이용하여 다양한 작업을 수행한다.

```python
from gensim.models import Word2Vec

model= Word2Vec.load("nvmx.model")
print("corpus_total_words: ", model.corpus_total_words)
print("-"*30)

print("-"*30)
print("사랑: ",model.wv['사랑']) #위에서 저장한 벡터 크기(200)만큼 단어"사랑"을 분산 저장한 것.
#즉, 200차원의 벡터에 단어의 의미를 분산하여 저장한 것이다. => model.wv 모든 수치를 불러온다.
print("-"*30)
```

```python
#단어의 유사도 계산
print("일요일= 월요일\t", model.wv.similarity(w1='일요일', w2='월요일'))
print("-"*30)

print("안성기=배우\t", model.wv.similarity(w1='안성기', w2='배우'))
print("-"*30)

print("대기업=삼성\t", model.wv.similarity(w1="대기업", w2="배우"))
print("-"*30)

print("일요일!=삼성\t", model.wv.similarity(w1="일요일", w2="삼성")) #수치가 60%이면 유사성이 멀다고 생각하면 된다.
print("-"*30)

print("히어로!=삼성\t", model.wv.similarity(w1="히어로", w2="삼성"))
print("-"*30)
```

```python
#가장 유사한 단어 추출
print(model.wv.most_similar("안성기", topn=5)) #가장 유사한 단어 5개 추출
print("-"*200)

print(model.wv.most_similar("시리즈", topn=5))
```
